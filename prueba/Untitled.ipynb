```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
I = cv2.imread('Pelota.png', cv2.IMREAD_GRAYSCALE)
```

```python
r, c = I.shape
def umbralizaNaive(I, um, bg = 0, fg = 255):
    O = np.zeros((I.shape), dtype = np.uint8)
    for i in range(r):
        for j in range(c):
            if I[i, j] >= um:
                0[i,j] = fg
            else:
                0[i,j] = bg
    return 0

def umbralIdealNaive(I):
    v = I.flatten()
    n = v.shape[0]
    O = np.zeros(v.shape)
    um = I[r//2, c//2]
    umOld = um
    cont = 0
    while True:
        mF = np.mean(v >= um)
        mB = np.mean(v < um)
        umOld = um
        um = (mF + mB) // 2
        if umOld == um or cont > 100:
            break
        cont += 1
    return um
```

```python
um = umbralIdealNaive(I)
print("El umbral es: ", um)

cv2.namedWindow("Original")
cv2.namedWidnow("Umbralizada")

cv2.imshow("Orignial", I)
#cv2.imshow("Umbralizada", O)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

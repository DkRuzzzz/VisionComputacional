```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
I = np.zeros((400, 400))

theta = np.pi/4

ct = np.cos(theta)
st = np.sin(theta)

xb = 200
yb = 200
T = np.array([[ct, -st, -ct * xb + st * yb + xb], [st, ct, -st * xb + ct *yb + yb], [0, 0, 1]])

for i in range (20, 40):
    for j in range (20, 75):
        cT = T @ np.array([j, i, 1]).T
        cT = np.round(cT / cT[2]).astype(np.int32)
        I[cT[1], cT[0]] = 1
        
#Display the image
fgl, axl = plt.subplots()
axl.imshow(I, cmap = 'Greys')
plt.show()
```

```python
crds = np.where(I != 0)

coors = np.hstack([crds[1],crds[0]])
xb = np.mean(coors[0,:])
yb = np.mean(coors[1,:])
r,c = coors.shape
n = c
coorsCentrales = coors - np.array([[xb], [yb]])
covMatrix = (coorsCentrales @ coorsCentrales.T) / (n-1)

# De manera equivalente
# covMatrix, np.cov(coorCentrales)

L, V = np.linalg.eig(covMatrix)
V, L
```
